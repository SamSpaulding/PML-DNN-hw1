```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sys
import os
import data_funcs
reload(data_funcs)
import neural_net

subjects = [1, 5, 7, 10, 15, 18, 25, 29, 35, 40]
filename = 'parkinsons_supervised.csv'

reload(neural_net)
neural_net.reload_files()
per_subject_arr = []
average_rmse = []
overall_rmse = []
mae_arr = []
for po in range(0, 5):
    net = neural_net.NeuralNetwork(filename, "homework")
    net.train(num_steps=100000, output_every_nth=99999)

    arr = net.test_on_test()
    per_subject_arr.append(arr)

    a = 0
    for i in range(0, len(arr)):
        if i+1 in subjects:
            a += arr[i]
    average_rmse.append(a/len(subjects))


    data_loader = data_funcs.DataLoader(filename)
    data = data_loader.data_dict
    def get_overall_rmse(data):
        sum1 = 0
        mae = 0
        total_points = 0
        for i in [1, 5, 7, 10, 15, 18, 25, 29, 35, 40]:
            sum1 = sum1 + (arr[i-1]**2)*len(data[i])
            mae = mae + ((arr[i-1]**2)*len(data[i]))**0.5
            total_points = total_points + len(data[i])
        sum1 = sum1 / total_points
        mae = mae/total_points
        sum1 = sum1 ** 0.5
        return sum1, mae
    overall_rmse.append(get_overall_rmse(data)[0])
    mae_arr.append(get_overall_rmse(data)[1])
```
